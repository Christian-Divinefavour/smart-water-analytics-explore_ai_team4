MODEL BUILING

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor , plot_importance, plot_tree
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df_Lake_Bilancino2.drop(['Lake_Level', 'Flow_Rate'],axis = 1)
y = df_Lake_Bilancino2[['Lake_Level', 'Flow_Rate']]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, shuffle = False)
X_train.set_index("Date", inplace = True)
X_test.set_index("Date", inplace = True)
```

```python
y_train_LakeLevel = y_train[['Lake_Level']]
y_train_FlowRate = y_train[['Flow_Rate']]
y_test_LakeLevel = y_test[['Lake_Level']]
y_test_FlowRate = y_test[['Flow_Rate']]
```

```python

m_LakeLevel = XGBRegressor()
m_LakeLevel.fit(X_train, y_train_LakeLevel)
```

```python
y_pred_LakeLevel = m_LakeLevel.predict(X_test)

score_mae = mean_absolute_error(y_test_LakeLevel, y_pred_LakeLevel)
score_rmse = math.sqrt(mean_squared_error(y_test_LakeLevel, y_pred_LakeLevel))


print('MAE  for Lake_Level using XGBRegressor is : {}'.format(score_mae))
print('RMSE for Lake_Level using XGBRegressor is : {}'.format(score_rmse))
```

```python
m_LakeLevel.get_booster().get_score()
```

```python
plot_importance(m_LakeLevel)
```

FLOW-RATE

```python

model_fr = XGBRegressor()
model_fr.fit(X_train, y_train_FlowRate)
```

```python
y_pred_FlowRate = model_fr.predict(X_test)
score_mae = mean_absolute_error(y_test_FlowRate, y_pred_FlowRate)
score_rmse = math.sqrt(mean_squared_error(y_test_FlowRate, y_pred_FlowRate))


print('MAE  for Flow_Rate using XGBRegressor is : {}'.format(score_mae))
print('RMSE for Flow_Rate using XGBRegressor is : {}'.format(score_rmse))
```

```python
model_fr.get_booster().get_score()
```

```python
plot_importance(model_fr)
```
