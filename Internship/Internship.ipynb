#### We look at the data statistics

We would like to see the variable statistics and interaction within each variable.

The following can be easily seen:

- There is a large difference between the max values and mean for most of the columns. This shows a lot of zero values collected during observation(the 2nd and 3rd percentiles is prove to this)

```python
#Checking the features with null values
null_values = df.isna().sum().to_frame().sort_values(0)
null_values.columns = ['Null']
def null_percentage(null):
    return round(null/8217*100,2)
null_values['Percentage_null'] = null_values.apply(null_percentage, axis=1)
provinces=list(null_values.index)
null_values
```

```python
#Visually comparing the percentage of null values in each column
ax=null_values.plot.barh(y='Percentage_null', rot=0, figsize=(16,14))
plt.title('Percentage NULL values for features', y=1, size=14)
```

We can see from the above table that the last 9 features have more than 40% null values. We'll be eliminating features with more than 40% null values

---

Kurtosis() help us detect outliers in our data. Typically, any feature with a kurtosis value > 10 will be investigated or worst case, removed. The above table contain the sorted kurtisis values of the features in our dataframe.

```python
df.kurtosis().to_frame().sort_values(0)
```

---
Owing to the seasonal nature of some of our features, we suspect that the reason for the high kurtosis values that was observed is because these features contain a significant fraction of zero values as can be seen below

```python
#Checking for zero observations in the features
lst = list(df.columns)
lst2 = []
for i in lst:
    lst2.append(df[df[i] == 0.000000].shape[0])
#zero_values=dict(zip(lst,lst2))
zero_values= {'Features':lst, 'Zero_Values':lst2}
zeros=pd.DataFrame(zero_values)
zeros.sort_values(by=['Zero_Values'])
```

```python
#Visualizing zero observations in the features
ax=zeros.plot.barh(y='Zero_Values', x= 'Features', rot=0, figsize=(16,14))
plt.title('Number of Zero values for features', y=1, size=14)
```

From the plot above, Rainfall_Cavallina has the highest 'zero' observations but from the null value plot, it has just 26.66% of it's total observations as NULL. This might imply that the region is in an arid area where rainfall is not just seasonal but sparse. We intend to confirm this by checking for seasonality in the time series data.
